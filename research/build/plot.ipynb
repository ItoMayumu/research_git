```python
import numpy as np
import matplotlib.pyplot as plt

with open("Temperature_result.vtk", "r") as f:
    lines = f.readlines()

data_start_idx = 0
for idx, line in enumerate(lines):
    if "LOOKUP_TABLE" in line:
        data_start_idx = idx + 1
        break

temperatures = np.array([float(line.strip()) for line in lines[data_start_idx:] if line.strip()])

Nx, Ny, Nz = 21, 21, 21
temp_3d = temperatures.reshape((Nx, Ny, Nz), order='F')

temp_2d = temp_3d[:, :, 10]

plt.figure(figsize=(6, 5))
contour = plt.contourf(temp_2d.T, levels=50, cmap="jet") # 画面の縦横を合わせるために転置(.T)
plt.colorbar(contour, label="Temperature [C]")
plt.title("2D Heat Diffusion Center Cross-Section (NumPy & Matplotlib)")
plt.xlabel("X Index")
plt.ylabel("Y Index")
plt.show()
```

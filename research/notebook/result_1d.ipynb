```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/result_1d.csv')

times_to_plot = [0.1, 0.5, 1.0]

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10,8), sharex=True)

for t in times_to_plot:
    df_t = df[df['Time'] >= t].groupby('x').first().reset_index()
    ax1.plot(df_t['x'], df_t['T'], label=f'Time = {t:.1f} s')
    ax2.plot(df_t['x'], df_t['alpha'], label=f'Time = {t:.1f} s')
    
ax1.set_ylabel('Temperature [°C]')
ax1.set_title('1D Boiling Simulation: Temperature & Liquid Fraction (Alpha)')
ax1.axhline(100, color='red', linestyle=':', alpha=0.7, label='T_sat (100°C)') # 沸点の補助線
ax1.legend()
ax1.grid(True)

# 下段グラフ（液相率）
ax2.set_xlabel('Position x [mm]')
ax2.set_ylabel('Alpha (1=Liquid, 0=Gas)')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```
